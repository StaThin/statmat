Esempio della Cauchy
--------
Per $n = 1, 2, \dots, 1000$, calcolo $\sum_{i=1}^n x_i/2$ dove $x_1, \dots, x_n$ sono Cauchy iid. Il campione casuale è ottenuto sapendo che la Cauchy è identica alla distribuzione $t_1$ con $1$ grado di libertà.

```matlab
rng(1200)
x = trnd(1,1000,1);    
```

```matlab
m = cumsum(x)./(1:1000)';
plot(1:1000, m, 'o')  % Figura 1.2
refline(0,0)
```
